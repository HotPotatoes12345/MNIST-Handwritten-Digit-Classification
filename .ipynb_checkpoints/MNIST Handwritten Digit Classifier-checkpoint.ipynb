```python
def initialize():
    '''
    Description
    This funciton initializes the parameters in the neural network with the W-weights initialized to small random
        numbers and the bias terms, b, intialized to zero
    
    Input
    layers_dim - the size of the neural network layers with the zero-th element being the number of examples
                - shape (1 x number of layers = L )
    
    Output
    parameters - a dictionary of the initialized parameters 
    
    '''
    
    np.random.seed(2) #sets the seed of the random number generator
    W = np.random.randn(1, X_train.shape[0])*0.01
    b = np.zeros((1,1))
        
    return W, b
    
```

```python
def sigmoid(x):
    '''
    Description
    The sigmoid function
    
    Input
    A value or array, X
    
    Ouput
    The sigmoid of the input
    '''
    
    sig = 1/(1+np.exp(-x))
    
    return sig
```

```python
sigmoid(-10)
```

```python
def deriv_sigmoid(x):
    '''
    Description
    The derivative of the sigmoid function
    
    '''
    
    d_sig = sigmoid(x) * (1 - sigmoid (x))
    
    return d_sig
```

```python
deriv_sigmoid(10)
```

```python
def forward_prop(W, b, X):
    
    Z = np.dot(W, X) + b
    A = sigmoid(Z)    
        
    return A, Z
```

```python
def propagate(X, W, b, Y):
    '''
    Description
    Calculates the forward pr|opogation steps of the neural network
    
    Input
    X_train   - the training examples
        shape >> (number of features, n, x number of examples, m )
    

    Output
    A_L    - the activation values of the final layer in the network
    caches - caches of all of the Z and A values of the hidden layers which will be used in back prop
    '''
    

    A, Z = forward_prop(W, b, X)
    
    m = A.shape[1]

    cost = -1/m * np.sum(Y * np.log(A) + (1 - Y) * np.log(1-A), axis=1) 
    

    dW = 2/m * np.dot(A - Y, X.T)
    db = 1/m * np.sum(A - Y, axis=1)

            
    return cost, dW, db
```

```python
def single_classifier_model(X_train, Y_train, learning_rate=0.1, iterations = 500, print_cost = True):
    '''
    Description
    Combines all aspects of the neural network into a single model

    Inputs
    
    
    Outputs
    '''
    
    costs = []
    W, b = initialize()
    
    for i in range(iterations):

        cost, dW, db = propagate(X_train, W, b, Y_train)
        
        W = W - learning_rate * dW
        b = b - learning_rate * db
        
        
        if i % 100 == 0:
            print('Cost for ' + str(i) + '= ' + str(cost) ) 
    
    return W, b, costs
```

```python
def error(X,W,b, y):
    '''
    Description
    Calculates the error of the predictions
    
    '''
    
    A, Z = forward_prop(W,b,X)
    
    predictions = A>0.5*1.0
    
    m = predictions.shape[1]
    
    error = 1/m * np.sum(predictions!=y)
    
    return error
```

```python
def all_classifier_model(X_train, y_train, learning_rate):
    '''
    Description
    This model calls the single_model fuction on every number in the data set (0-9)
    
    '''
    
    parameters = {}
#   cost_dic = {}
    m = X_train.shape[1]
    
    for i in range(10):
        y_train_temp = y_train ==i
        W, b, costs = single_classifier_model(X_train, y_train_temp, learning_rate)
        parameters['W_'+str(i)] = W
        parameters['b_'+str(i)] = b
#        costs['costs_'+str(i)] = costs

    return parameters
```

```python
all_parameters = all_classifier_model(X_train, y_train, learning_rate=0.1)
```

```python
def all_predict(parameters, X):
    
    A_values, Z = forward_prop(parameters['W_'+str(0)], parameters['b_'+str(0)], X)
    for i in range(1,10):
        A_temp, Z = forward_prop(parameters['W_'+str(i)], parameters['b_'+str(i)], X)
        A_values = np.append(A_values, A_temp, axis=0)
    
    predictions = []
    for j in range(X.shape[1]):
        predictions.append(np.argmax(A_values[:,i]))
    
    return predictions
```
